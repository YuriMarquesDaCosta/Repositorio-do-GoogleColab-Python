##Letra a). Carregar uma imagem e transformá-la em um matriz de valores reais. (2,0 pontos)


```python
import numpy as np
import cv2
from skimage import io
from google.colab.patches import cv2_imshow

origem = "https://www.fightersgeneration.com/news2021/char4/ryu-by-kinu-nishimura-snk-vs-capcom-poster.jpg" #link da imagem
image = io.imread(origem) 
img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY) #transformando em preto e branco

cv2_imshow(img)
```

```python
from PIL import Image 
from numpy import asarray

numpydata = asarray(img) 
print(type(numpydata)) 
print("Matriz da imagem:")
print(numpydata.shape)
```

```python
print("cada pixel da imagem:")
numpydata = asarray(img) 
print(numpydata)
```

##Letra b) Aplicar a Decomposição da matriz por Valores Singulares. (2,0 pontos)


```python
from numpy.linalg import svd
U, Singular, V = svd(img)
print("U: ",U)
print("-------------------------------")
print("Valores Singulares:", Singular)
```

##c. Manter apenas os valores que representam 80% dos Valores Singulares. (2,0 pontos).

```python
print("Valores Singulares a 80%:", (Singular/0.8)) #80% da imagem 
```

##d. Apresentar a imagem transformada com a aplicação do item (c).

```python
img = (Singular/0.8)
print("Imagem com 80%: ") #A diferença é leve mas existe
cv2_imshow(img) 
```
